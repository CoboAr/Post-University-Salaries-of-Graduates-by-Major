Challenge: The Highest Mid-Career Salary


```python
print(clean_df['Mid-Career Median Salary'].max())
print(f"Index for the max mid career salary: {clean_df['Mid-Career Median Salary'].idxmax()}")
clean_df['Undergraduate Major'][8]
```

Challenge: The Lowest Starting Salary.



```python
print(clean_df['Starting Median Salary'].min())
print(f"Index for the min starting career salary: {clean_df['Starting Median Salary'].idxmin()}")
clean_df['Undergraduate Major'][49]
```

Challenge: The Lowest Mid-Career Salary

```python
clean_df.loc[clean_df['Mid-Career Median Salary'].idxmin()]
```

```python
spread_col= clean_df['Mid-Career 90th Percentile Salary']- clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1,'Spread',spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

Challenge: Find the degrees with the highest potential.

```python
high_90_percentile=clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_90_percentile[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

Challenge: Majors with the Greatest Spread in Salaries

```python
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

Challenge: Majors with the Greatest Mid-Career Median Salary

```python
highest_mid_career = clean_df.sort_values('Mid-Career Median Salary', ascending=False)
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```
